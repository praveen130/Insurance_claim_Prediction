```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR = Q3-Q1
    lower_range = Q1-(1.5*IQR)
    upper_range = Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
lowprice,uppprice = remove_outlier(train['charges'])
train['charges']=np.where(train['charges']>uppprice,uppprice,train['charges'])
train['charges']=np.where(train['charges']<lowprice,lowprice,train['charges'])
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
factors = ['age','sex','bmi','children','smoker','region','charges']
```

```python
clf = DecisionTreeClassifier()
X = data[factors]
test_data = data_test[factors]
y = train.insuranceclaim
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
clf.fit(X, y)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
